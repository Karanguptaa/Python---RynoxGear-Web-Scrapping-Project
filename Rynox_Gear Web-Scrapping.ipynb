# ACCESS URL

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = "https://rynoxgear.com/collections/jackets/products/rynox-stealth-evo-3-jacket-black-grey?variant=12466579144727"
```

```python
r = requests.get(url)
htmlcontent = r.content
htmlcontent
```

# Parse the HTML Link

```python
soup = BeautifulSoup(htmlcontent,"html.parser")
#print(soup)
soup.prettify #Prettify the codes of the HTMl page
```

# ACCESS NAME OF PRODUCT

```python
title = soup.title
print(type(soup)) # BeautifulSoup
print(type(title)) # Tag
print(type(title.string)) # Navigablestring
```

```python
title=soup.title
title
```

# ACCESS PRICE

```python
Price = soup.find_all("div",class_="selection-wrapper price product-single__price-product-template")
Price
```

```python
Product_Price = []
for i in range(0,len(Price)):
    Product_Price.append(Price[i].get_text())
Product_Price
```

# Product Features

```python
feature = soup.find("div",class_="product_features").text
feature.strip().replace('\n\n\n\n\n','   ')
```

# ACCESS CUSTOMERS NAMES

```python
customers = soup.find_all("span",class_="jdgm-rev__author-wrapper")
customers
```

```python
cust_name = []
for i in range(0,len(customers)):
    cust_name.append(customers[i].get_text())
cust_name
```

# ACCESS REVIEW

```python
Comment = soup.find_all("div",class_="jdgm-rev__body")
Comment
```

```python
Review_Comment = []
for i in range(0,len(Comment)):
    Review_Comment.append(Comment[i].get_text())
Review_Comment
```

# ACCESS RATINGS

```python
rating_one = soup.find("span",class_="jdgm-rev__rating")['data-score']
rating_one
```

```python
ratings = [span['aria-label']for span in soup.find_all("span",class_="jdgm-rev__rating")]
ratings
```

# ARRANGE ALL DATA USING PANDAS LIBRARY

```python
import pandas as pd
df = pd.DataFrame({"Customers_NAme": cust_name,
                   "Review_Comments": Review_Comment,
                   "Review_Stars": ratings})
print(df)
```

```python
df.info()
```

# Export data to CSV file

```python
df.to_csv("Rynox_Gear_reviews.csv",index = False)
```
